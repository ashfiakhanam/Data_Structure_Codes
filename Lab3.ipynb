```python
class Node:
    def __init__(self, value = None, next = None, prev = None):
        self.value = value
        self.next = next
        self.prev = prev

class DoublyList: 
    head = Node()
    def __init__ (self, a = None):
        if type(a) == list:
            self.tail = self.head
            for i in range(len(a)):
                self.tail.next = Node(a[i], self.head, self.tail)
                self.tail = self.tail.next
                self.head.prev = self.tail
        else:
            print("array cannot be empty")

    def showList(self):
        node1 = self.head.next
        res = ""
        while node1 != self.head:
            res += str(node1.value)+", "
            node1 = node1.next
        print(res)

    def insertAtTail(self, newElement):
        node1 = self.head.prev
        node2 = Node(newElement, self.head, node1)
        node1.next = node2
        self.head.prev = node2   

    def insert(self, newElement, index):   
        size = self.getSize()
        if size == 0:
            return
        if index > size+1 or index <= 0:
            print("Invalid Index")
            return
        res = self.ValueCheck(newElement)
        if res == False:
            print("Value already exists")
            return

        if index == size+1:
            self.insertAtTail(newElement)
            return

        if index <= (size/2 + size%2):
            c = 0
            node1 = self.head
            while c <= index:
                if c == index:
                    node2 = Node(newElement, node1.next, node1)
                    node1.next.prev = node2
                    node1.next = node2
                    return
                node1 = node1.next
                c += 1
        else:
            c = size
            node1 = self.head
            while c >= index:
                if c == index:
                    node2 = Node(newElement, node1.next, node1)
                    node1.prev.next = node2
                    node1.next = node2
                    return
                node1 = node1.prev
                c -= 1    

    def getSize(self):
        if(self.head.next == None):
            print("The List is Empty!")
            return 0
        c = 1
        node1 = self.head.next
        while (node1.next.value != None):
            c += 1
            node1 = node1.next
        return c

    def ValueCheck(self, newElement):
        node1 = self.head.next
        while(node1.value != None):
            if(node1.value == newElement):
                return False
            node1 = node1.next
        return True

    def remove(self, index):
        size = self.getSize()
        if size == 0:
            return

        if index > size or index <= 0:
            print("Invalid Index")
            return

        if index <= (size/2 + size%2):
            c = 0
            node1 = self.head
            while(c <= index):
                if(c == index):
                    node1 = node1.next
                    node1.next.prev = node1.prev
                    node1.prev.next = node1.next
                    node1 = None
                    return
                node1 = node1.next
                c += 1
        else:
            c = size
            node1 = self.head
            while c >= index:
                if c == index:
                    node1 = node1.prev
                    node1.next.prev = node1.prev
                    node1.prev.next = node1.next
                    node1 = None
                    return          
                node1 = node1.prev
                c -= 1    

    def removeKey(self, key):
        size = self.getSize()
        if size == 0:
            return

        res = self.ValueCheck(key)
        if res == True:
            print("Value doesn't exist")
            return
        halfSize = (size/2 + size%2)
        node1 = self.head
        cf= 1
        while cf<= halfSize:
            if node1.value == key:
                node1.next.prev = node1.prev
                node1.prev.next = node1.next
                node1 = None
                return key
            node1 = node1.next
            cf+= 1
        node1 = self.head
        c = size
        while c >= halfSize:
            if node1.value == key:
                node1.next.prev = node1.prev
                node1.prev.next = node1.next
                node1 = None
                return key
            node1 = node1.prev  
            c -= 1
```
