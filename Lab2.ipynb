```python
class  Node:
    value = None
    next = None 

class MyList():
    def __init__ (self, list0 = None):
        self.head = None
        self.temp = None
        if list0 == None :
            self.head = None

        elif type(list0) == list :
            for i in range(len(list0)):
                node1 = Node()
                if i != 0:
                    node1.value = list0[i]
                    node1.next = None
                    self.temp.next = node1
                    self.temp = node1

                else:
                    node1.value = list0[i]  
                    node1.next = None
                    self.head = node1
                    self.temp = node1  


        elif type(list0)== MyList:
            self.head = list0.head
     
    def showList(self):
        node1 = self.head
        if node1 == None:
            print("empty list")
        while node1!= None:
            print(node1.value, end = ", ")
            node1 = node1.next
        print(" ")

    def isEmpty(self):
        li = self.head
        if li == None:
            return True
        else:
            return False

    def clear(self):
        node1 = self.head
        while node1 != None:
            self.head = node1.next
            node1 = None
            node1 = self.head

    def insert(self, newElement):
        target = Node()
        target.value = newElement
        node1 = Node()
        node1 = self.head
        while node1.next != None:
            if node1.next == target:
                print("key exists!")
                break
            else:
                node1 = node1.next
        node1.next = target
        return

    def insertAt(self, newElement, index):
        node = Node()
        node.value = newElement
        node.next = None

        if index == 0:
            node.next = self.head
            self.head = node     
        else:
            temp = self.head
            for i in range(index-1):
                temp = temp.next
            node.next = temp.next
            temp.next = node

    def remove(self, deleteKey):
        if self.head == None:
            return
        node1 = Node()
        node1 = self.head
        while node1.next != None:
            if node1.value == deleteKey:
                deleted = node1.value
                node1.value = None
                return deleted
            node1 = node1.next
        
    def evenChecker(self):
        array = []
        node1 = Node()
        node1 = self.head
        while node1.next != None:
            if node1.value % 2 == 0:
                array.append(node1.value)
            node1 = node1.next
        if node1.value % 2 == 0:
                array.append(node1.value)        
        evenList = MyList(array)
        evenList.showList()

    def elementChecker(self, element):
        node1 = Node()
        node1 = self.head
        while node1.next != None:
            if node1.value == element:
                print(True)
                return ""
            node1 = node1.next
        print(False)

    def reversedList(self):
        start = self.head
        head = None
        while start != None:
            temp = start.next
            start.next = head
            head = start
            start = temp
        self.head = head
        self.showList()

    def sortedList(self):
        x = self.head
        while x != None:
            y = x.next
            while y != None:
                if y.value < x.value:
                    temp = x.value
                    x.value = y.value
                    y.value = temp
                y = y.next
            x = x.next
        self.showList()

    def sumList(self):
        sum = 0
        current = Node()
        current = self.head
        while current != None :
            sum += current.value
            current = current.next
        print(sum)

    def rotate(self, direction, k):
        length = 1
        c = Node()
        c = self.head
        while c.next != None:
            length += 1
            c = c.next 
        if c.next == None:
            c.next = self.head

        if direction == "right":
            i = 0
            temp = Node()
            tail = Node()
            temp = self.head
            while i <= length-k:
                if i == length-k-1:
                    tail = temp

                if i == length-k:
                    self.head = temp
                temp = temp.next
                i += 1
            temp = temp.next
            tail.next = None

        elif direction == "left":
            i = 0
            temp = Node()
            temp = self.head
            while i <= k-1:
                if i == k-1:
                    self.head = temp.next
                    temp.next = None
                    break
                temp = temp.next
                i += 1
        self.showList() 
```
